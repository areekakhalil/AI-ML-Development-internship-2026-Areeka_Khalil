```python
import pandas as pd
```

```python
df=pd.read_csv("Titanic-Dataset.csv")
```

```python
print("Shape of DataSet")
print(df.shape)
```

```python
print("Column Name")
print(df.columns)
```

```python
print("DataType")
print(df.dtypes)
```

```python
print("Fiest Five Row of Dataframe")
print(df.head)
```

```python
print("Dataset information")
print(df.info)
```

```python
print("Missing Values in Columns ")
print(df.isnull().sum())
```

```python
survived_count=df['Survived'].sum()
print("Number OF passenger Servived")
print(survived_count)
```

```python
average_age=df['Age'].mean()
print("Average Age is:",average_age)

max_age=df['Age'].max()
print("Maximum Age is:",max_age)

min_age=df['Age'].min()
print("Minimum Age is:",min_age)
```

```python
print("Shape before Cleaning:")
print(df.shape)
```

```python
df_clean=df.dropna(subset=['Age'])
```

```python
print("Shape After Cleaning:")
print(df_clean.shape)
```

```python
remove_rows=df.shape[0]-df_clean.shape[0]
print("Numbers ofRemove_Rows:")
print(remove_rows)
```

```python
df_clean.to_csv("titanic_cleaned.csv", index=False)

print("\nCleaned dataset saved successfully as 'Titanic-Dataset.csv'")
```
