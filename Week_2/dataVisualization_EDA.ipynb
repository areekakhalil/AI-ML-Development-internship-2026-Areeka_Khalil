```python
#import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
#load DataSet
from google.colab import files
uploaded=files.upload()
df=pd.read_csv("Titanic-Dataset (3).csv")
df.head()
```

```python
plt.figure(figsize=(7,5))

sns.countplot(
    x='Survived',
    hue='Sex',
    data=df,
    palette='Set1'
)

plt.title("Survival Count base on  Gender")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")

plt.show()
```



Female passengers had higher survial rate as compare to male.



```python
#histogram
plt.figure(figsize=(7,5))

plt.hist(
    df['Age'].dropna(),
    bins=20,
    color='teal',
    edgecolor='white'
)

plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")

plt.show()
```

Most Passengers are between 30 to 40 mean they are adults

```python
#box plot
plt.figure(figsize=(7,5))

sns.boxplot(
    x='Pclass',
    y='Fare',
    data=df,
    palette='Set2'
)

plt.title("Ticket Fare by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare Paid")

plt.show()
```

The passenger in first class paid the highest fares and widest rane of ticket price.

```python
#Heat Map
numeric_df = df[['Age', 'Fare', 'Pclass', 'Survived', 'SibSp', 'Parch']]
corr = numeric_df.corr()

plt.figure(figsize=(8,6))

sns.heatmap(
    corr,
    annot=True,
    fmt='.2f',
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```


Pclass and Fare have the strongest negative correlation. This means passengers in higher classes (lower class number) generally paid higher ticket fares.
